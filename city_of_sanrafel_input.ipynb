```python
import requests
from bs4 import BeautifulSoup

link ="https://www.cityofsanrafael.org/major-planning-projects-2/"
res= requests.get(link)
html=res.text

soup = BeautifulSoup(res.text, 'html.parser')
text=soup.find('div',class_='panel-widget-style panel-widget-style-for-55564-0-0-2')
print(text)
table=text.find('table', class_='table')
print(table.text)



```

```python
import pandas as pd

df = pd.read_html(str(table))[0]
!mkdir Excel
df.to_excel('./Excel/city_Of_Sanrafael_output_file.xlsx', index=False)
```
